```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from keras import layers
from keras.layers import Input, Dense, Activation, Flatten
from keras.initializers import glorot_uniform
from keras.regularizers import l2
from keras.models import Model
from keras.utils import layer_utils
```

```python
url = 'https://raw.githubusercontent.com/MubashirullahD/Tweets-Sentiment-Analysis/master/tweet_sentiment.csv'
df = pd.read_csv(url)
df.drop(df.columns[[0]], axis = 1, inplace = True) 
```

```python

features = df.drop(['Positive','Negative', 'Neutral'], axis = 1)
target  =  df.loc[:, ['Positive', 'Negative', 'Neutral']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size = 0.1, random_state = 42,
                                                    stratify = target)   
```

```python
df
```

```python
def SentimentModel(input_shape=(200,)):
    """
    Implementation of the SentimentModel.
    
    Arguments:
    input_shape -- shape of the tweets in the dataset

    Returns:
    model -- a Model() instance in Keras
    """
     
    X_input = Input(input_shape)
    
    X = Dense(15, activation='relu', kernel_regularizer=l2(0.01), name='fc1', kernel_initializer = glorot_uniform(seed=0))(X_input)
    
    X = Dense(3, activation='softmax', kernel_regularizer=l2(0.01), name='fc2', kernel_initializer = glorot_uniform(seed=0))(X)
    
    model = Model(inputs = X_input, outputs=X, name='SentimentModel')
    
    return model
```

```python
SentimentModel = SentimentModel()
```

```python
SentimentModel.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
SentimentModel.fit(x=X_train, y=Y_train, epochs=400, batch_size=64)
```

```python
preds = SentimentModel.evaluate(x=X_test, y=Y_test)
### END CODE HERE ###
print()
print ("Loss = " + str(preds[0]))
print ("Test Accuracy = " + str(preds[1]))
```
